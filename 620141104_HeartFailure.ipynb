```python
df
```

```python
y = df['HeartDisease'] #Assign Heart Disease value to y since it is our target variable
```

```python
df.info()
```

```python
df=df.drop(columns=['HeartDisease','Cholesterol','MaxHR','RestingECG_LVH','RestingECG_Normal','RestingECG_ST'])
```

```python
df.info()
```

```python
x = df.values
```

```python
y = y.values
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest =train_test_split(x, y, test_size=0.2, random_state=42) #split data into training and test data for our model
```

```python
xtrain.shape 
```

```python
xtest.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
xtrain = sc.fit_transform(xtrain) #scale the data
```

```python
xtrain
```

```python
xtest= sc.transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import classification_report
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
cnf_matrix = metrics.confusion_matrix(ytest, ypred)
print(cnf_matrix)
print(classification_report(ytest, ypred))
```

```python
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier(max_depth=5, random_state=42, criterion="gini", splitter='best')
model2.fit(xtrain, ytrain)
```

```python
ypred = model2.predict(xtest)
cnf_matrix = metrics.confusion_matrix(ytest, ypred)
print(cnf_matrix)
print(classification_report(ytest, ypred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier(n_jobs=-1, n_estimators=100, random_state=42)
model3.fit(xtrain, ytrain)
```

```python
ypred = model3.predict(xtest)
cnf_matrix = metrics.confusion_matrix(ytest, ypred)
print(cnf_matrix)
print(classification_report(ytest, ypred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn_model = KNeighborsClassifier()
kn_model.fit (xtrain, ytrain)
preds = kn_model.predict(xtest)
print(classification_report(ytest, preds))
cnf_matrix2 = metrics.confusion_matrix(ytest, preds)
print(cnf_matrix2)
```

```python
from sklearn.neural_network import MLPClassifier
mlp_model = MLPClassifier(hidden_layer_sizes=(100), batch_size= 42, learning_rate= 'constant', activation = 'tanh', alpha= 0.04,learning_rate_init=0.004)
mlp_model.fit(xtrain, ytrain)
```

```python
preds2 = mlp_model.predict(xtest)
print(classification_report(ytest, preds2))
cnf_matrix2 = metrics.confusion_matrix(ytest, preds2)
print(cnf_matrix2)
```

```python
from sklearn import svm
smodel = svm.SVC( kernel='rbf')
smodel.fit(xtrain, ytrain)
preds2 = smodel.predict(xtest)
print(classification_report(ytest, preds2))
cnf_matrix2 = metrics.confusion_matrix(ytest, preds2)
print(cnf_matrix2)
```
